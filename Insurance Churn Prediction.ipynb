So, the dataset is highly imbalanced. Apart from the below work, I have also attempted model building using sampled data (over sampling) but those models were found to be poorer perfomers than what I got here

```python
#Comparing the models
compare_models()
```

Because the evaluation criteria of this hackathon is based on F1 score, we can see the top performers are lightgbm, catboost, gbc and xgboost. Thus, let's create these four models

```python
#Creating the catboost model
catboost = create_model('catboost')
```

```python
#Creating the lightgbm model
lgbm = create_model('lightgbm')
```

```python
#Creating the gradient boosting model
gbc = create_model('gbc')
```

```python
#creating the xgboost model
xgboost = create_model('xgboost')
```

So, depending on the mean F1 scores, lightgbm is the best one. Now, lets stack these model and check whether we can see any improvement

```python
stacker = stack_models(estimator_list = [lgbm,xgboost,gbc], meta_model = catboost)
```

```python
stacker1 = stack_models(estimator_list = [lgbm,catboost,gbc], meta_model = xgboost)
```

```python
stacker2 = stack_models(estimator_list = [lgbm,catboost,xgboost], meta_model = gbc)
```

```python
stacker3 = stack_models(estimator_list = [gbc,catboost,xgboost], meta_model = lgbm)
```

Now let's check which stacker performes best in the unseen holdout data

```python
pred_holdout = predict_model(stacker)
```

```python
pred_holdout = predict_model(stacker1)
```

```python
pred_holdout = predict_model(stacker2)
```

```python
pred_holdout = predict_model(stacker3)
```

So, we can see 'stacker' is the best performer with the unseen data. Thus, let's finalize this stacked model and make a prediction on the test data

```python
model_final = finalize_model(stacker)
```

```python
predictions = predict_model(model_final, data = test)
```
